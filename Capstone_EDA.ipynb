### Seed song data

Notes:

- The numeric columns are not scaled, which will be problematic when building models

```python
#remove object data types to keep only numerical columns
df = df_seed.select_dtypes(exclude=['object'])
df.shape
```

```python
# plot all numerical columns at once
col_names = df.columns
plot_num = 1

plt.subplots(7,2, figsize=(20,15))

for col in col_names:
    plt.subplot(7,2,plot_num)
    sns.histplot(df[col])
    plot_num +=1

plt.tight_layout()
plt.show()
```

```python
# heat map
fig,ax = plt.subplots(1,1,figsize=(20,20))
sns.heatmap(df.corr(),
            cmap='coolwarm', # The best color palette I saw for this data, others have too much color 
            annot=True, #add numbers
            mask = np.triu(df.corr()), #removed the redundant half of the heatmap
            fmt=".01f", #this might be controversial because I am rounding...
            # ... but the trends do not change when I add more digits; it just makes it more crowded.
            linewidth=.5) #added some extra flare to improve readability
```

Notes:

- Heatmap shows interesting correlations:
    - strong positive correlations between: loudness and energy, liveness and acousticness
    - strong negative correlations between: speechiness and danceability, tempo and danceability, duration and danceability
- The numeric columns are not scaled, which will be problematic when building models
- All my selected seed songs have a time signature of 4.

### I want to map my data over the dataset

```python
# check column name to see overlap
print(df.columns)
print(ds_numeric_only.columns)
```

```python
# both dataframes have these column names in common
common_columns = ['danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness',
                  'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo',
                  'duration_ms', 'time_signature']

plt.subplots(5, 3, figsize=(20, 15))

for plot_num, col in enumerate(common_columns, start=1):
    plt.subplot(5, 3, plot_num)
    sns.histplot(ds_numeric_only[col], color='blue', alpha=0.5, label='Spotify dataset')
    sns.histplot(df[col], color='orange', alpha=0.5, label='Seed song data')

    # Add vertical lines for mean and standard deviation of Seed song data
    mean_value = df[col].mean()
    std_value = df[col].std()

    plt.axvline(mean_value, color='red', linestyle='dashed', linewidth=2, label='Mean')
    plt.axvline(mean_value + std_value, color='green', linestyle='dashed', linewidth=2, label='Mean + 1 SD')
    plt.axvline(mean_value - std_value, color='green', linestyle='dashed', linewidth=2, label='Mean - 1 SD')

    plt.legend()

plt.tight_layout()
plt.show()
```

Notes:

- I cannot see my seed song data in the figures as the count is too low.
- I added a vertical line to represent mean and standard deviation, which help very much.
- From some of the standard deviations of my seed song data, I should reconsider picking different songs to lower it and hopefully improve clustering.
