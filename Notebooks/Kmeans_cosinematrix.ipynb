### Making bad bunny dataset (seed) from the same dataset

```python
import pandas as pd

# Assuming you have a DataFrame spotify_df containing the entire Spotify dataset
spotify_df = pd.read_csv('../spotify-datasets/kaggle-yamaerenay-dataset/tracks.csv')

# Check if 'Bad Bunny' is in the 'artists' column for each row
bad_bunny_condition = spotify_df['artists'].apply(lambda x: 'Bad Bunny' in eval(x))

# Create a new dataset containing only rows where 'Bad Bunny' is true
bad_bunny_dataset = spotify_df[bad_bunny_condition]

# Save the new dataset to a CSV file
#bad_bunny_dataset.to_csv("bad_bunny_dataset.csv", index=False)

# Print the number of occurrences of 'Bad Bunny'
count_bad_bunny = bad_bunny_condition.sum()
print(f"Number of occurrences of 'Bad Bunny': {count_bad_bunny}")
```

### K means algorithm

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

# Assuming you have a DataFrame df containing your 50 favorite songs and API metadata
df = bad_bunny_dataset

# Extract relevant features and create a feature matrix
feature_columns = ['danceability', 'energy', 'tempo']  
feature_matrix = df[feature_columns].values

# Normalize the feature matrix
scaler = StandardScaler()
normalized_feature_matrix = scaler.fit_transform(feature_matrix)

# Define the pipeline with PCA and KMeans
pipeline = Pipeline([
    ('pca', PCA()),
    ('kmeans', KMeans())
])

# Specify the hyperparameter grid for GridSearchCV
param_grid = {
    'pca__n_components': [2, 3, 4, 5],  # Adjust the number of components as needed
    'kmeans__n_clusters': range(2, 15)  # Adjust the range of clusters
}

# Create GridSearchCV
grid_search = GridSearchCV(pipeline, param_grid, cv=5, scoring='neg_mean_squared_error')

# Fit the pipeline to find the best settings
grid_search.fit(normalized_feature_matrix)

# Get the best parameters
best_params = grid_search.best_params_
print("Best Parameters:", best_params)

# Get the best estimator (pipeline)
best_pipeline = grid_search.best_estimator_

# Use the best pipeline to transform your seed data and perform clustering
cluster_labels = best_pipeline.named_steps['kmeans'].fit_predict(normalized_feature_matrix)

# Add the cluster labels to your seed DataFrame
df['cluster'] = cluster_labels

# Assuming you have a DataFrame spotify_df containing the entire Spotify dataset
spotify_df = pd.read_csv('../spotify-datasets/kaggle-yamaerenay-dataset/tracks.csv')

# Extract relevant features from the Spotify dataset
spotify_feature_matrix = spotify_df[feature_columns].values

# Normalize the Spotify feature matrix
normalized_spotify_feature_matrix = scaler.transform(spotify_feature_matrix)

# Predict clusters for Spotify songs using the trained k-means model
spotify_cluster_labels = best_pipeline.named_steps['kmeans'].predict(normalized_spotify_feature_matrix)

# Add the predicted cluster labels to the Spotify DataFrame
spotify_df['predicted_cluster'] = spotify_cluster_labels

# Filter the Spotify DataFrame to get recommended songs in the same cluster as your seed data
recommended_songs = spotify_df[spotify_df['predicted_cluster'] == cluster_labels[0]]

# Save the recommended songs to a CSV file
recommended_songs.to_csv("../output/recommended_songs.csv", index=False)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

# Assuming you have a DataFrame df containing your 50 favorite songs and API metadata
df = bad_bunny_dataset

# Extract relevant features and create a feature matrix
feature_columns = ['danceability', 'energy', 'tempo']  
feature_matrix = df[feature_columns].values

# Normalize the feature matrix
scaler = StandardScaler()
normalized_feature_matrix = scaler.fit_transform(feature_matrix)

# Define the pipeline with PCA and KMeans
pipeline = Pipeline([
    ('pca', PCA(n_components=3)),  # Adjust the number of components as needed
    ('kmeans', KMeans(n_clusters=10, random_state=42))  # Adjust the number of clusters as needed
])

# Use the pipeline to transform your seed data and perform clustering
cluster_labels = pipeline.named_steps['kmeans'].fit_predict(normalized_feature_matrix)

# Add the cluster labels to your seed DataFrame
df['cluster'] = cluster_labels

# Assuming you have a DataFrame spotify_df containing the entire Spotify dataset
spotify_df = pd.read_csv('../spotify-datasets/kaggle-yamaerenay-dataset/tracks.csv')

# Extract relevant features from the Spotify dataset
spotify_feature_matrix = spotify_df[feature_columns].values

# Normalize the Spotify feature matrix
normalized_spotify_feature_matrix = scaler.transform(spotify_feature_matrix)

# Predict clusters for Spotify songs using the trained k-means model
spotify_cluster_labels = pipeline.named_steps['kmeans'].predict(normalized_spotify_feature_matrix)

# Add the predicted cluster labels to the Spotify DataFrame
spotify_df['predicted_cluster'] = spotify_cluster_labels

# Use cosine similarity to find similar songs
seed_cluster_center = pipeline.named_steps['kmeans'].cluster_centers_[cluster_labels[0]]
cosine_similarities = cosine_similarity([seed_cluster_center], normalized_spotify_feature_matrix)

# Get indices of top 50 similar songs
top_50_indices = np.argsort(cosine_similarities[0])[::-1][:50]

# Extract the top 50 recommended songs from the Spotify dataset
recommended_songs_cosine = spotify_df.iloc[top_50_indices]

# Save the recommended songs to a CSV file
recommended_songs_cosine.to_csv("../output/recommended_songs_cosine.csv", index=False)
```
